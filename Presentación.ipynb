

## “SELECCIÓN DE RUTAS AÉREAS, EN BASE A LAS DISTANCIAS DE LOS ENTRE AEROPUERTOS DEL PERÚ MEDIANTE EL ALGORITMO DE KRUSKAL Y DIJKSTRA”




## 1. DESCRIPCIÓN DEL PROBLEMA

Debido a la necesidad de poder viajar a las principales ciudades del país, muchas personas toman la decisión de ir vía terrestre, aéreo, marítimo o fluvial en algunos casos. De trasportarse vía terrestre se enfrentan a diversos problemas de llegar en muchas horas, el mal estado de las vías terrestres por diversos factores, ya sea por el factor de precipitaciones o mal mantenimiento. Solo algunos lugares de la selva realizan el transportarse por vía fluvial esto es imposible la mayoría de los casos. 

Debido a esto, el planteamiento de integrar a todos los aeropuertos del país y que haya servicios de transporte aéreo eficaz, es decir en menor tiempo y a un precio accesible de la población.

Para ello se necesita identificar una ruta aérea principal que integré a todos los aeropuertos del país, el cual tenga una mínima distancia. A partir de esto identificar cuáles serán las rutas mínimas de un aeropuerto a otro cualquiera. 


## 2. OBJETIVOS

### Objetivo General
- Identificar las rutas aéreas más cortos, con la condición de integrar todos los aeropuertos del Perú.

### Objetivos Específicos

- Descubrir la ruta más corta de un aeropuerto ‘X’ a otro ‘Y’ en Perú.
- Identificar una ruta aérea con menor distancia posible que integre a todos los aeropuertos del país.
- Medir la dimensión en la que se va reducir las rutas aéreas al incluir rutas alternas a la ruta principal integrada.

## 3.3 La Ruta Aérea más Corta que integre a todos los aeropuertos(Ruta Aérea Principal)

Con el objetivo de encontrar una ruta aérea principal que integre a los 28 aeropuertos del Perú, vamos hacer uso del algoritmo de Kruskal.


### Creamos el algoritmo de Krukal

Este algoritmo se ha desarrollado con el lenguaje de programación Python, en 4 etapas, la primera de contiene una funcion para hacer una lista conjuntos de nodos, la segunda es una función para encontrar el nodo que aún no está conectado, la tercera es una función de unir con el objetivo de añadir al nodo vecino con menor distancia y finalmente la función principal de Kruskal que integra los tres funciones anteriores y que va tener una entrada de dato de tipo disccionario.

```python

Nodo = dict() #Creamos un diccionario.
            #Un diccionario es una colección desordenada, modificable e indexada.
    
resultado = {} ## se crea un diccionario vacio para resultado
conjuntos = []  ## se crea una lista vacia de conjuntos

# 1.CREAMOS UNA FUNCION PARA HACER CONJUNTO NODOS 
def Hacer_conjunto(vertice):  
    Nodo[vertice] = vertice
    
# 2.CREAMOS UNA FUNCIÓN PARA ENCONTRAR EL NODO QUE AUN NO ESTÁ CONECTADO
def Encontrar_conjunto(vertice):
    if Nodo[vertice] != vertice:
        Nodo[vertice] = Encontrar_conjunto(Nodo[vertice])
    return Nodo[vertice]

# 3. CREAMOS UNA FUNCIÓN PARA UNIR NODOS CON LA MENOR ARISTA

def Union(u, v, Ordenada):
    
    if u not in conjuntos:
        conjuntos.append(u)
    
    if v not in conjuntos:
        conjuntos.append(v)
        
    print("Conjuntos unidos:", conjuntos)

    Dato1 = Encontrar_conjunto(u)
    Dato2 = Encontrar_conjunto(v)
    if Dato1 != Dato2:
        for Dato in Ordenada:
            Nodo[Dato1] = Dato2
            
# 4. LA FUNCIÓN PRINCIPAL DEL ALGORITMO DE KRUSKAL

def Kruskal(grafo):
    resultante = []
    cont = 0
    for vertice in grafo['A']:
        Hacer_conjunto(vertice)
    
    Ordenada = list(grafo['B']) ## Hacemos lista de los arcos 
    Ordenada.sort() ## Ordena la lista de arcos de menor a mayor
    Ordenada = [(a,b,c) for c,a,b in Ordenada] ## cambio de posición (A,B,5)que el valor se muestre al final
    print("==============================")
    print("DATOS ORDENADOS:")
    print("==============================")
    print("ORDENADOS:", Ordenada)
    Ordenada = [(c,a,b) for a,b,c in Ordenada] # vuelve a la misma orden (5,A,B)
    
    for Dato in Ordenada:
        peso, u, v = Dato # peso:distancia; u y v son los puntos o vertices
        
        if Encontrar_conjunto(u) != Encontrar_conjunto(v):
            resultante.append(Dato)
            print("==============================")
            print("PASO Nro.", cont)
            print("==============================")
            resultante = [(a,b,c) for c,a,b in resultante]
            print("RESULTANTE : ", resultante)
            resultante = [(c,a,b) for a,b,c in resultante]
            cont+=1
            Union(u, v, Ordenada)

    return resultante
```
